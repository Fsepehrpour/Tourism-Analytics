```python
# Get unique regions and define a color map
regions = data_cleaned['Entity'].unique()
colors = cm.get_cmap('tab10', len(regions))  # Use a colormap for distinct colors
```

```python
# Loop through regions and assign specific colors and line styles
line_styles = ['-', '--', '-.', ':']  # Define a set of line styles
for idx, region in enumerate(regions):
    region_data = data_cleaned[data_cleaned['Entity'] == region]
    plt.plot(
        region_data['Year'], 
        region_data['International tourist arrivals by region'], 
        label=region, 
        color=colors(idx),  # Assign a specific color
        linestyle=line_styles[idx % len(line_styles)],  # Assign a line style
        linewidth=2  # Adjust line thickness
    )

# Customize the plot
plt.title('Trend Analysis of International Tourist Arrivals by Region', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Arrivals', fontsize=12)
plt.grid()

# Add the legend outside the plot
plt.legend(title='Region', fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```

### Conclusion

1. Regional Dominance:
    - Europe consistently recorded the highest number of international tourist arrivals from 1995 to around 2019. This suggests that Europe has been the most popular destination globally, likely due to its rich cultural heritage, historical landmarks, and diverse attractions.
    - East Asia and the Pacific showed significant growth, ranking as the second-highest region in tourist arrivals, demonstrating its rising popularity as a tourism hub.

2. Steady Growth Trends:
    - Regions like Africa, Americas, and South Asia displayed steady growth trends over the years, albeit at a slower pace compared to Europe and East Asia. These regions likely benefit from increasing infrastructure development and promotion of tourism.

3. The Impact of 2020:
    - The year 2020 shows a sharp decline across all regions, indicative of the severe impact of the COVID-19 pandemic on global tourism. This highlights the vulnerability of the tourism industry to global crises.

4. Middle East and South Asia:
    - Middle East and South Asia maintained modest but consistent growth in tourist arrivals, reflecting developing tourism industries that are gradually gaining global interest.

5. The Role of the "Not Classified" Category:
    - The "Not classified" category remains relatively low and stable, likely representing either incomplete data or niche regions not officially categorized.

6. Year-on-Year Growth Patterns:
    - From 1995 to 2019, most regions exhibited consistent year-on-year growth, suggesting a globally increasing interest in travel and tourism. Notable exceptions include slight dips during economic or political disruptions (e.g., the financial crisis of 2008–2009).
